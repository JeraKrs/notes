## 1.2 什么是操作系统

* 从用户角度出发，操作系统是一个控制软件，它管理应用程序，并为应用程序提供服务，有时负责终止程序。
* 从系统内部出发，操作系统负责资源的管理，它管理外部设备，为进程分配资源，并将CPU、磁盘、内存抽象成进程、文件、地址空间。

**操作系统的架层次结构**：操作系统在硬件之上，应用程序之下。它分为两个部分，外壳（Shell）和内核（Kernel）。

* Shell：用户使用这个程序来操作计算机系统进行工作，它一般以图形界面（GUI）和命令行形式存在。
* Kernel：操作系统的核心部分，负责管理CPU、内存和磁盘等重要硬件。它的主要任务有CPU的调度、物理内存的管理、虚拟内存的管理、文件系统的管理、中断处理与设备驱动。

**操作系统内核的特征**：
* 并发：同时存在多个运行的程序。
* 共享：多个应用程序共享计算机资源，有”同时“共享和互斥共享两种方式。
* 虚拟：理由多道程序设计技术，让每个用户觉得有一个计算机专门为自己服务。
* 异步：程序的执行不是连续一贯的，但只要运行环境相同，操作系统需要让程序的运行结果相同。

> TIP: 并发和并行
> * 并发：一段时间内有多个程序在运行。
> * 并行：一个时刻内有多个程序在运行（多核处理器）。

## 1.6 操作系统的历史

1. 早期计算机使用纸带传输程序和数据，操作系统只起到加载的作用。
2. CPU等硬件快速发展，计算机速度得到提升，性能未得到充分利用。
3. 成批、离线处理。
4. 内存的容量越来越大，通过多道程序设计让CPU可以同时执行多个程序。
5. 分时系统的出现，依赖外部设备：时钟。
6. 个人电脑操作系统。
7. 网络的快速发展，出现了分布式的操作系统。

## 1.7 操作系统的结构


**宏内核**：整个核心程序都是以核心空间（Kernel Space）的身份及监管者模式（Supervisor Mode）来运行。

**微内核**：尽可能把内核功能移动到用户空间。内核只负责中断处理、消息传递等核心功能，而像文件系统、内存管理、网络协议栈等则由外部进程负责。这样做的代价则是系统开销变大。

**外内核**：外内核的设计分为两部分，一部分与应用程序紧密结合，另一部则抽象硬件。

**虚拟机**：虚拟机（VMs）是由虚拟机监视器（VMM）虚拟出一台完整的计算机，它的操作系统则运行在VMM之上，目标是多操作系统共享硬件资源。
